```python
import pandas as pd


test_data = cg.get_coin_market_chart_by_id(id='bitcoin',vs_currency='usd',days='50')
prices = test_data['prices']
volume = test_data['total_volumes']
mcap = test_data['market_caps']


time_list = []
price_list = []

for value in prices:
    time_list.append(value[0])
    price_list.append(value[1])

df = pd.DataFrame({'time':time_list, 'price':price_list})
    
```
