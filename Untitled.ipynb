```python
import numpy as np 
from tqdm import tqdm 
from binance.client import Client 
import pandas as pd
from datetime import datetime
import random

import talib

import plotly_express as px
```

```python
#select cryptocurrencies you'd like to gather and time interval
ratios = ['BTC']
START_TIME = '21 May, 2020'
END_TIME = '21 May, 2021'
api_key=''
api_secret=''
client = Client(api_key=api_key,api_secret=api_secret)
```

```python
# merge = False
for ratio in ratios:
    print(f'Gathering {ratio} data...')
    data = client.get_historical_klines(symbol=f'{ratio}USDT',interval=Client.KLINE_INTERVAL_1MINUTE,start_str=START_TIME,end_str=END_TIME)
    cols = ['time','open','high','low','close','volume','CloseTime','QuoteAssetVolume','NumberOfTrades','TBBAV','TBQAV','null']
    df = pd.DataFrame(data,columns=cols)
    print('complete')
```
